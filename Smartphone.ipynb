```python
# Read the smartphone data (adjust the file path as needed)
sm = pd.read_csv('smartphones.csv')   # load data into a DataFrame
```

```python
# Extract the 'rating' column and drop missing values (if any)
rating_data = sm['rating'].dropna()   # It is just to ensure there are no NaNs for plotting
```

#### from the boxplot, i don't see any outliers, that means i can do mean or median on rating column

```python
# OR fill with mean
sm['rating'] = sm['rating'].fillna(sm['rating'].mean())
```

```python
sm.isnull().sum()
```

```python
# imputing missing values on caltegorical columns "OS" , "CARD" with mode
sm["os"]=sm["os"].fillna(sm["os"].mode()[0])
sm["card"]=sm["card"].fillna(sm["card"].mode()[0])
```

```python
sm.isnull().sum()
```

```python
sm=sm.dropna(subset=["camera"])
```

### String Manipulation

```python
sm['price']=sm['price'].str.replace("₹","").str.replace(",","").astype(int)
```

```python
sm['ram_clean'] = sm['ram'].str.extract(r'(\d+)\s*GB\s*RAM', expand=False).astype(float)
```

```python
sm['rom_clean'] = sm['ram'].str.extract(r'(\d+)\s*GB\s*inbuilt', expand=False).astype(float)
```

```python
sm['battery_clean'] = sm['battery'].str.extract(r'(\d+)\s*mAh').astype(float)
```

```python
sm['display_clean'] = sm['display'].str.extract(r'(\d+\.\d)\s*inches').astype(float)
```

```python
sm["os_clean"]=sm["os"].str.extract(r'Android v(\d+)').astype(float)
```

```python
sm['processor_speed_GHz'] = sm['processor'].str.extract(r'(\d+\.\d)\s*GHz').astype(float)
# since preocessor is kinda complicated , i only took it's clockspeed
```

```python
sm['primary_camera_MP'] = sm['camera'].str.extract(r'(\d+)\s*MP').astype(float)
# i just took the first camera
```

#### Recleaning the dataset

```python
columns_to_clean = [
    'ram_clean',
    'rom_clean',
    'battery_clean',
    'display_clean',
    'os_clean',
    'processor_speed_GHz',
    'primary_camera_MP'
]

sm_imp = sm.copy()

for column in columns_to_clean:

    Q1 = sm_imp[column].quantile(0.25)
    Q3 = sm_imp[column].quantile(0.75)
    IQR = Q3 - Q1
        
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
        
    # For replace values outside the bounds with lower bound & upper bound
    sm_imp[column] = sm_imp[column].clip(lower_bound, upper_bound)

    
    sm_imp[column].fillna(sm_imp[column].median(), inplace=True)

# Print the size of the final DataFrame
print(f"Original number of rows: {len(sm)}")
print(f"Number of rows after filling and capping: {len(sm_imp)}")
```
